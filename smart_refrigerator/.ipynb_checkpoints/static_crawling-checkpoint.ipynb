```python
import requests
from bs4 import BeautifulSoup
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from pyvirtualdisplay import Display
from selenium.webdriver.chrome.options import Options

# 시작번호 https://www.10000recipe.com/recipe/6947317

start = time.time()

food_list = []
ingredient_list = []
seasoning_list = []
how_to_list = []
recipe_num_list = []

recipe_num = 6947317
check_num = 1;

while True:
    req = requests.get("https://www.10000recipe.com/recipe/"+ str(recipe_num))
    html = req.text
    if len(html) < 100:
        while True:
            recipe_num -= 1;
            req = requests.get("https://www.10000recipe.com/recipe/"+ str(recipe_num))
            html = req.text
            if len(html) > 200:
                break;
        
    bs = BeautifulSoup(html, 'html.parser')
    
    print(check_num, ':', recipe_num, sep='', end=', ')
    # 음식이름
    d_food = bs.select('#contents_area > div.view2_summary.st3 > h3')
    food = [dom.text for dom in d_food]
    food_str = ''.join(food)
    food = []
    food_list.append(food_str)

    # 재료
    d_ingredient = bs.select('#divConfirmedMaterialArea > ul:nth-of-type(1)')
    ingredient_str = [dom.text for dom in d_ingredient]
    ingredient_str = ''.join(ingredient_str).split()[1:]
    ingredient_list.append(ingredient_str)
    ingredient_str = []



    # 양념
    d_seasoning = bs.select('#divConfirmedMaterialArea > ul:nth-of-type(2)')
    seasoning_str = [dom.text for dom in d_seasoning]
    seasoning_str = ''.join(seasoning_str).split()[1:]
    seasoning_list.append(seasoning_str)
    seasoning_str = []


    # 조리방법
    d_howto = bs.select('.view_step')
    hlist = [dom.text for dom in d_howto][0].replace('원본보기', '').replace('\n', '')
    how_to_str = hlist.split("\r")[0].split('\xa0')
    how_to_str = [a for a in how_to_str if a]
    how_to_list.append(how_to_str)
    hlist = []
    
    recipe_num_list.append(recipe_num)
    
    recipe_num -= 1;
    check_num += 1;
    if recipe_num == 6947217:
        break;
    
df = pd.DataFrame({"name" : food_list,
                   "ingredient" : ingredient_list,
                   "seasoning" : seasoning_list,
                   "howto" : how_to_list,
                   "recipe_num" : recipe_num_list})  

df.to_csv("./output/static_test1.csv", index=False)
        
print("time: ", time.time()-start)
```

```python
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from pyvirtualdisplay import Display
from selenium.webdriver.chrome.options import Options
chrome_options=Options()
display = Display(visible=0, size=(1024, 768))
display.start()

options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disalbe-gpu')
driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver', options=options)
driver.implicitly_wait(4) 
driver.get('https://www.10000recipe.com/recipe/list.html?q=&query=&cat1=&cat2=&cat3=70&cat4=&fct=&order=reco&lastcate=cat3&dsearch=&copyshot=&scrap=&degree=&portion=&time=&niresource=')
time.sleep(2)


start = time.time()

recipe_list = []
main_list = []
food_list = []
food = []
ingredient_list = []
ingredient = []
ingredient_name_list = []
ingredient_name = []
seasoning_list = []
seasoning = []
seasoning_name_list =[]
seasoning_name = []
how_to_list = []
hlist = []

check_num = 1;
recipe_num = 1;
pnum = 2;
# 첫 1~10페이지 체크용
page = 1; 

while True:    
    url = driver.current_url

    # 레시피 선택
    select_recipe = driver.find_element_by_css_selector('ul > ul > li:nth-child('+ str(recipe_num) +') > div.common_sp_thumb > a > img')
    if not select_recipe:
        break;
    select_recipe.click()
    print(recipe_num, end=', ')
#     time.sleep(1)
    
    # 음식이름
    name_list = driver.find_elements_by_css_selector(' div.view2_summary.st3 > h3')
    food = [name.text for name in name_list]

    # 재료
    ing_list = driver.find_elements_by_css_selector('#divConfirmedMaterialArea > ul:nth-child(1)')
    ingredient_str = [name.text for name in ing_list]
    ingredient_str = ''.join(ingredient_str).split('\n')[1:]

    # 양념
    ssn_list = driver.find_elements_by_css_selector('#divConfirmedMaterialArea > ul:nth-child(2)')
    seasoning_str = [name.text for name in ssn_list]
    seasoning_str = ''.join(seasoning_str).split('\n')[1:]

    # 조리 방법
    how_to = driver.find_elements_by_css_selector('.view_step')
    hlist = [name.text for name in how_to][0]
    
    food_str = ''.join(food)
    food = []
    food_list.append(food_str)

    ingredient_list.append(ingredient_str)
    ingredient_str = []
    
    seasoning_list.append(seasoning_str)
    seasoning_str = []
    
    how_to_str = hlist.split("\n레시피와 관련된 상품")[0].split('\n')
    hlist = []
    how_to_list.append(how_to_str)
    
    recipe_num += 1;
    except_list = [6, 15, 26, 35]
    if recipe_num in except_list :
        recipe_num += 1;
    driver.back()
    driver.get(url)
    
    # 마지막 레시피 체크
    if recipe_num == 45:
        print('page: ', page)
        # 페이지 이동
        if page == 100:
            break;
        next_page = driver.find_element_by_css_selector("#contents_area_full > ul > nav > ul > li:nth-child("+ str(pnum) +") > a")
        ppnum = pnum + 1;
        page_check = driver.find_elements_by_css_selector("#contents_area_full > ul > nav > ul > li:nth-child("+ str(ppnum) +") > a")
        if page_check :
            next_page.click()
            page += 1;
#             time.sleep(1)
            pnum += 1;
            recipe_num = 1;
        else :
            if pnum == 12 :
                next_page.click()
                page += 1;
#                 time.sleep(1)
                pnum = 3;
                recipe_num = 1;
            else :
                if page > 12:
                    break;


df = pd.DataFrame({"food" : food_list,
                   "ingredient" : ingredient_list,
                   "seasoning" : seasoning_list,
                   "howto" : how_to_list})                    
# df = pd.DataFrame({"food" : food_list,
#                    "ingredient" : ingredient_list,
#                    "ingredint_name" : ingredient_name_list,
#                    "seasoning" : seasoning_list,                    
#                    "seasoning_name" : seasoning_name_list,
#                    "howto" : how_to_list})

df.to_csv("./output/소고기레시피.csv", index=False)
        
driver.quit()
print("time: ", time.time()-start)
```
