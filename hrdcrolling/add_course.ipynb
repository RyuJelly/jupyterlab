## 03_금융보험

```python
import pandas as pd
from konlpy.tag import Okt
from konlpy.utils import pprint
import nltk
from konlpy.tag import Hannanum 
import re
df = pd.read_csv("03_금융보험.csv")
df.dropna(subset=['취업률'], inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
indexlist = []
for i in range(len(df)):
    indexlist.append(i)
# df1 = df.reindex(indexlist) <- 이거 왜 안 됌?
df['index'] = indexlist
df = df.set_index('index')
df['NCS분야'] = "03_금융보험"

df_c = df['과정명'].copy()
df_a = df_c.tolist()
df_a = '\n'.join(df_a)
df_a = re.sub('양성', '', df_a)
df_a = re.sub('실무자', '', df_a)
# t = Okt()
# tokens_ko = t.nouns(df_a)

# ko_ = nltk.Text(tokens_ko)  
# ko_.vocab()
# freq_dic = dict(ko_.vocab())
# print(freq_dic)

hannanum = Hannanum()
    
t = Okt()
tokens_ko = hannanum.nouns(df_a)


ko_ = nltk.Text(tokens_ko)  
ko_.vocab()
freq_dic = dict(ko_.vocab())
[freq_dic.pop(key) for key in ['합격', '과정', '총무', '비즈니스반']]
sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)
sortedDict = sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)

print(sortedDict)

sortlist = []
for i in range(len(sortedDict)) :
    sortlist.append(sortedDict[i][0])
print(sortlist)
```

```python
courselist = []
for i in range(len(df)) :
#     df_c[i] = re.sub("['양성', '실무자']", '', df_c[i])
    df_c[i] = re.sub('양성', '', df_c[i])
    df_c[i] = re.sub('실무자', '', df_c[i])
    tokens = hannanum.nouns(df_c[i])
    tx = nltk.Text(tokens)
    freq_dic = dict(tx.vocab())
    [freq_dic.pop(key, None) for key in ['합격', '과정', '총무', '비즈니스반']]
    fre = list(freq_dic.keys())
    fre = fre[0]
    if fre in sortlist:
        courselist.append(fre)
    else:
        courselist.append('None')

df['과정'] = courselist
display(df)

df.to_csv("preprocessing/03_금융보험.csv")
```
