```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from konlpy.utils import pprint
import nltk    
from wordcloud import WordCloud,STOPWORDS           
import matplotlib.pyplot as plt 

# 폰트 설정
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 기존에 "[1] 리뷰내용" 형식으로 저장해놓아서 데이터프레임 -> 리스트변환 -> 넘버링제거 -> 문자열변환 
# 과정을 거쳤는데 다하고보니 '명사'만 추출할 것이므로 그냥 read만 했어도 상관없었겠구나
# 라고 생각했을땐 이미 늦어버려서 그대로 썼습니다.

df_hotelreview = pd.read_table('data/naverhotel.txt', header=None)

# 넘버링 제거
df_delnum = df_hotelreview[0].str[4:]

# 리스트로 만들어 사용하려했으나 konlpy 오류 발생
list_hotelreview = list(np.array(df_delnum.tolist()))

# konlpy는 문자열밖에 처리 못하므로 다시 문자열 변환
str_hotelreview = '\n'.join(list_hotelreview)

# 트위터 형태소 분석기 사용
t = Okt()
tokens_ko = t.nouns(str_hotelreview) # 명사 토큰만 추출

# 추출한 토큰으로 명사 객체 생성
ko_ = nltk.Text(tokens_ko)       

# 빈도수로 추출 후 딕셔너리로 저장
freq_word = ko_.vocab()
freq_dic = dict(freq_word)

# generate_from_frequencies에서는 stopwords 지원하지 않음.
# 직접 빼줘야 한다.
freq_dic.pop('호텔')

# ---사용 불가----
# stopwords = set()                               
# stopwords.add('호텔')

myfontpath = "data/THEdog.ttf" 

wordcloud = WordCloud(font_path = myfontpath,
                      background_color = 'white',
                      width = 700, height = 500,
                      colormap='inferno').generate_from_frequencies(freq_dic)

wordcloud.to_file('output/python_hw.png')

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
