```python
from selenium import webdriver
driver = webdriver.Chrome('c:/Temp/chromedriver')
driver.implicitly_wait(6) 
driver.get('http://gs25.gsretail.com/gscvs/ko/products/event-goods')
import re
import time
time.sleep(4)

goodsname = []
goodsprice = []

# 2+1 클릭
twotoone = driver.find_element_by_css_selector('#TWO_TO_ONE')
twotoone.click()
time.sleep(2)

# 마지막 페이지 번호
end_page = driver.find_element_by_css_selector('div > div:nth-child(5) > div > a.next2')
end_page_num = int(re.sub("[^0-9]", '', str(end_page.get_attribute('onclick'))))

while True:    
    # 상품명 추출
    name_list = driver.find_elements_by_css_selector('div:nth-child(5) > ul > li > div > p.tit')
    for name in name_list:
        goodsname.append(name.text)
    
    # 상품가격 추출
    price_list = driver.find_elements_by_css_selector('div:nth-child(5) > ul > li > div > p.price > span')
    for price in price_list:
        p = re.sub("[^0-9]", '', price.text)
        goodsprice.append(p)
    
    # 마지막 페이지 체크
    page = driver.find_element_by_css_selector('div > div:nth-child(5) > div > span > a.on')
    page_num = int(page.text)
    if page_num == end_page_num:
        break;
    # 페이지 이동
    next_page = driver.find_element_by_css_selector('div > div:nth-child(5) > div > a.next')
    next_page.click()
    time.sleep(1)
    
with open('C:/PyStexam/exercise/gs25_twotoone.csv', 'wt', encoding='utf-8') as f:
    f.write('goodsname,goodsprice\n')  
    for i in range(len(goodsname)):
        f.write(goodsname[i]+","+goodsprice[i]+'\n')  
        
driver.quit()
```
