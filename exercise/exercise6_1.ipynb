```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

font_path = '../data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_csv("../data/cctv_seoul.csv", index_col='구별')
df = df['CCTV수']
mycolor = sns.color_palette('Spectral', len(df))

plt.figure(figsize=(10,10))
df.plot(kind='barh', width=0.7, color=mycolor)
plt.title("시각화 과제(1) 각 구의 CCTV 설치 현황", size=25)
plt.xlabel('각 구에 설치된 CCTV 댓수', size=20)
plt.ylabel('구이름', size=20)
plt.xticks(size=12)
plt.yticks(size=12)
plt.grid()

plt.savefig('../output/hw1.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns
mycolor = sns.color_palette('PRGn', 26)

font_path = '../data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_csv("../data/cctv_seoul.csv", index_col='구별')
df.sort_values(by='CCTV수', ascending=False, inplace=True)
df = df['CCTV수']
df = df.T

df.plot(kind='bar', figsize=(20,10), width=0.7, color=mycolor)
plt.title("시각화 과제(2) 각 구의 CCTV 설치 현황", size=25)
plt.ylabel('각 구에 설치된 CCTV 댓수', size=20)
plt.xlabel('구이름', size=20)
plt.xticks(size=12, rotation=20)
plt.yticks(size=12)
plt.grid()

plt.savefig('../output/hw2.png')
plt.show()
```
