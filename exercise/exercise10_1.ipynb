```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
from konlpy.utils import pprint
import nltk    
from wordcloud import WordCloud,STOPWORDS           
import matplotlib.pyplot as plt 

# 폰트 설정
from matplotlib import font_manager, rc
font_path = "../data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

df = pd.read_table("../exercise/naverhotel.txt", header=None)
df_a = df[0]
df_a = df_a.tolist()
df_a = '\n'.join(df_a)

t = Okt()
tokens_ko = t.nouns(df_a)
print(tokens_ko)

ko_ = nltk.Text(tokens_ko)  
print(ko_)
freq_dic = dict(ko_.vocab())
freq_dic.pop('호텔')

myfontpath = "data/THEdog.ttf" 

wordcloud = WordCloud(font_path = myfontpath,
                      background_color = 'white',
                      width = 700, height = 500,
                      colormap='inferno').generate_from_frequencies(freq_dic)

wordcloud.to_file('../output/python_hw.png')

fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
