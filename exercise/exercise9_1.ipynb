```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import seaborn as sns

font_path = '../data/malgun.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 1 - (1)
df = pd.read_csv('../data/product_click.log', sep=' ', header=None)
df.columns = ['clickdate', '상품ID']
df1 = df.groupby('상품ID')
df2 = df1.count()
df2.columns = ['클릭횟수']
df2.sort_values(by='클릭횟수', ascending=False, inplace=True)
df2 = df2['클릭횟수']
mycolor = sns.color_palette('Spectral', len(df2))
df2.plot(kind='bar', figsize=(20,20), width=0.7, color= mycolor)
plt.xticks(rotation=20)
plt.xlabel('상품 ID', size=20)
plt.ylabel('클릭횟수', size=20)
plt.grid()
plt.show()
```
