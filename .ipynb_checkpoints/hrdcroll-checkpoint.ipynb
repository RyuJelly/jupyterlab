```python
from selenium import webdriver
import time, re
import pandas as pd

driver = webdriver.Chrome('C:/Temp/chromedriver')
driver.implicitly_wait(3) 

# gs25 페이지로 이동
driver.get('http://www.hrd.go.kr/hrdp/ti/ptiao/PTIAO0100L.do')
time.sleep(2)
```

```python
# 내일배움, 국가기간전략, 4차산업, 청년취업 4항목만  

courseList = []
academyList = []
locateList1 = []
locateList2 = []
termList = []
priceList = []
limitList = []
emprateList = []
ageList = []
lastcourse = ''
count = 0

countdom = driver.find_element_by_xpath('//*[@id="searchForm1"]/div[3]/p/span')
counttext = countdom.text
maxcount = int(re.sub('[,]', '', counttext))

print('전체갯수 : ', maxcount)
while True:
    # 페이지 만큼 반복
    nextbutton = driver.find_elements_by_xpath('//*[@id="searchForm1"]/div[4]/a[contains(@class,"next")]')
    
    # > 버튼이 없으면 버튼 인덱스 3 ~ 12
    # > 버튼이 없으면 버튼 인덱스 1 ~ 10
    if(len(nextbutton) != 0):
        startnum = 3
    else:
        startnum = 1
        
    for b_num in range(startnum,startnum+10):
        
        courselist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/span[1]/p/a')
        academylist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/dl/dt/a')
        locatelist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/dl/dd')
        termlist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/ul[2]/li[1]')
        pricelist = driver.find_elements_by_class_name('price')
        limitlist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/ul[1]/li[5]')
        empratelist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/ul[1]/li[2]')
        agelist = driver.find_elements_by_xpath('//*[@id="searchForm1"]/ul/li/div[1]/ul[2]/li[3]')
            
        for i in range(len(courselist)):
            course = courselist[i].text
            # 학원명
            academy = academylist[i].text
            # 도, 시
            locatetext = locatelist[i].text
            fix_locatetext = re.sub('[()☎-]', '', locatetext)
            fix_locatetext = re.sub('[0-9]', '', fix_locatetext.strip())
            sp_locatetext = fix_locatetext.split(' ')
            if(len(sp_locatetext) < 2):
                locate1 = None
                locate2 = None
            else:  
                locate1 = sp_locatetext[0]
                locate2 = sp_locatetext[1]
            # 기간
            termtext = termlist[i].text
            s_termtext = termtext.split(',')
            term = re.sub('[^0-9]','',s_termtext[0])
                # 훈련비
            pricetext = pricelist[i].text
            price = re.sub('[^0-9]','',pricetext)
                # 정원
            limittext = limitlist[i].text
            limit = re.sub('[^0-9]','',limittext)
                # 취업률
            empratetext = empratelist[i].text
            emprate = re.sub('[가-힣\n :%]', '', empratetext)
                # 연령대
            agetext = agelist[i].text
            age = re.sub('[^0-9]', '', agetext)
                
                # 리스트에 저장
            courseList.append(course)
            academyList.append(academy)
            locateList1.append(locate1)
            locateList2.append(locate2)
            termList.append(term)
            priceList.append(price)
            limitList.append(limit)
            emprateList.append(emprate)
            ageList.append(age)
                
            count += 1                
            print(count, end = ",")
                
                # 다음 페이지 이동(1)
        print(b_num, '버튼넘버의 페이지 로드 끝')
            
        if(maxcount == count):   
            print('for문 종료')
            break
        # 마지막 페이지이면 for문 빠져나와 넥스트 버튼 클릭
        if(b_num == 12):
            break;
        else:
            numbutton = driver.find_elements_by_xpath('//*[@id="searchForm1"]/div[4]/a['+str(b_num)+']')
            # 다음 넘버 버튼 없으면 스탑
            if(len(numbutton) != 0):
                numbutton[0].click()
                time.sleep(5)
            else:
                break
            
    if(maxcount == count):
        print('while 종료')
        break
        
    # 1 - 10 페이지 이동 후 다음 페이지 
    nextbutton = driver.find_elements_by_xpath('//*[@id="searchForm1"]/div[4]/a[contains(@class,"next")]')
    if(len(nextbutton) != 0):
        nextbutton[0].click()
        time.sleep(5)
    else:
        break
    
    
df = pd.DataFrame({'과정명':courseList, '학원명':academyList, '기간':termList,
                   '훈련비':priceList, '정원':limitList, '취업률':emprateList,
                   '연령대':ageList, '지역1':locateList1, '지역2':locateList2})
display(df)

df.to_csv("./hrdcrolling/10_영업판매.csv")

```
