```python
import pandas as pd
import seaborn as sns
import numpy as np
iris = sns.load_dataset("iris")
iris_x = iris.loc[:,['sepal_length', 'sepal_width',
            'petal_length', 'petal_width']]
import random
random.seed(1)
for col in range(4):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('nan')
# iris_x.head(10)

print("[1번] NaN이 포함된 모든 행 제거")
iris_x_1 = iris_x.dropna(axis=0)
display(iris_x_1.head())

print("[2번] NaN이 3개 이상 포함된 행 제거")
iris_x_2 = iris_x.dropna(axis=0, thresh=3)
display(iris_x_2.head(10))

print("[3번] sepal_length,sepal_width열에 NaN이 포함된 행 제거")
iris_x_3 = iris_x.dropna(subset=['sepal_length','sepal_width'], how='any', axis=0)  
display(iris_x_3.head(10))

print("[4번] 모든 NaN의 값 0으로 대체")
iris_x_4 = iris_x.fillna(0, inplace=False)
display(iris_x_4.head(10))

print("[5번] 모든 NaN의 값 평균값으로 대체")
mean_value = iris_x.mean(axis=0)
# print(mean_value)
iris_x_5 = iris_x.fillna(round(mean_value, 1), inplace=False)
display(iris_x_5.head(10))

print("[6번] 모든 NaN의 값 10으로 대체")
iris_x_6 = iris_x.fillna(10, inplace=False)
display(iris_x_6.head(10))

print("[7번] 모든 NaN의  값 이전 행의 값으로 대체")
iris_x_7 = iris_x.fillna(method='ffill', inplace=False)
display(iris_x_7.head(10))
```

```python
# 3-1번
df_clicklog = pd.read_csv('data/product_click.log', header=None, sep=' ')
print(df_clicklog.head(5))
print(df_clicklog.info())
```

```python
# 3-2번
df_clicklog.columns = ['oldclicktime','clickpid']
print(df_clicklog.head(5))
```

```python
# 3-3번
df_clicklog.oldclicktime = df_clicklog.oldclicktime.astype('object') 
print(df_clicklog.info())
```

```python
# 3-4번
df_clicklog['newclicktime'] = pd.to_datetime(df_clicklog.oldclicktime , format='%Y%m%d%H%M')

print(df_clicklog.info())
print(df_clicklog.head(5))
```

```python
# 3-5번
df_clicklog['year'] = df_clicklog['newclicktime'].dt.year
df_clicklog['month'] = df_clicklog['newclicktime'].dt.month
df_clicklog['day'] = df_clicklog['newclicktime'].dt.day
df_clicklog['hour'] = df_clicklog['newclicktime'].dt.hour
df_clicklog['minute'] = df_clicklog['newclicktime'].dt.minute
print(df_clicklog.info())
print(df_clicklog.head(5))
```
