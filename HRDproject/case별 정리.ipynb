## 20대 취업률 70%이상 과정

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('hrdcrolling/preprocessing/final.csv', index_col = 0)
df1 = df.copy()
df1.dropna(subset=['연령대'], inplace=True)
df1 = df1.groupby(["과정"]).mean()
df1.sort_values(by=['취업률'], ascending=False, inplace=True)
df1 = df1.loc[(df1.연령대 < 30) & (df1.연령대 >= 20), :]
df70 = df1['취업률'] >= 70
df1 = df1[df70]
print('취업률 70% 이상인 과정의 수: ', len(df1))
display(df1)


fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1, 1, 1)

sns.barplot(x=df1.index, y='취업률', data=df1, palette='rocket', ax=ax1)
plt.xticks(size=20, rotation=10)
plt.yticks(size=20)
plt.grid(True)
plt.ylim(60, 100)

plt.savefig("20대_취업률_70%이상_과정.png")
plt.show()
plt.close()
```

## 20대 선호 과정 TOP15 (과정수기준)

```python
import matplotlib.pyplot as plt
import seaborn as sns

df2 = df.copy()
df2 = df2.loc[(df.연령대 < 30) & (df.연령대 >= 20), :]
df2.dropna(subset=['연령대'], inplace=True)
df2.rename(columns = {'학원명' : '과정수'}, inplace = True)
df2 = df2.groupby(["과정"]).count()
df2.sort_values(by='과정수', ascending=False, inplace=True)
top70 = int(len(df2)*0.3)
df2 = df2.head(top70)
display(df2)

# 구역 설정
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1, 1, 1)

# 그래프 그리기
sns.barplot(x=df2.index, y='과정수' ,data=df2, palette="rocket", ax=ax1)
plt.xticks(size=15, rotation=20)
plt.yticks(size=20)
plt.grid(True)

# 출력 
plt.savefig("20대선호과정top70%.png")
plt.show()
plt.close()
```

```python
scorelist = range(40,40-len(df1),-1)
df1 = df1.loc[:,['취업률']]
df1['취업률점수'] = scorelist
display(df1)

scorelist = range(40,40-len(df2),-1)
df2 = df2.loc[:,['과정수']]
df2['선호도점수'] = scorelist
display(df2)

df_20_best = pd.concat([df1, df2], axis=1, join='inner')
df_20_best['최적점수'] = df1['취업률점수'] + df2['선호도점수']
df_20_best = df_20_best.sort_values(by=['최적점수'], ascending=False)
display(df_20_best)

fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(1,1,1)

sns.barplot(x=df_20_best.index, y='취업률', data=df_20_best, palette='rocket', ax=ax1)
plt.xticks(size=20, rotation=10)
plt.yticks(size=20)
plt.ylim(70,78)
plt.grid(True)
ax2 = plt.twinx()
sns.lineplot(x=df_20_best.index, y='과정수' ,data=df_20_best, ls='--', color='blue', ax=ax2)
plt.yticks(size=20)

plt.savefig('output/20대최적점수과정.png')
plt.show()
plt.close()
```
