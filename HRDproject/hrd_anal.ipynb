## 1. 과정별 취업률 통계

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 전처리
df_Bycourse = df.loc[:,['NCS','취업률']]
grouped = df_Bycourse.groupby(['NCS'])
df_Bycourse = grouped.mean()
df_Bycourse.sort_values(by=['취업률'], ascending=False, inplace=True)

# 구역 설정
fig = plt.figure(figsize=(20, 10))
ax1 = fig.add_subplot(1, 1, 1)

# 그래프 그리기
sns.barplot(x=df_Bycourse.index, y="취업률", data=df_Bycourse, palette="rocket", ax=ax1)
plt.xticks(size=14, rotation=70)
plt.yticks(size=14)
plt.grid(True)
plt.ylim(20, 75)

# 텍스트 설정

# 출력 
plt.show()
```

## 2. 학원기준 과정의 개수(전문성) & 취업률

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 전처리
df_academy_course = df.loc[:,['학원명','과정']]

a = df_academy_course.drop_duplicates(subset = ['학원명','과정'])
grouped = a.groupby(['학원명'])
df_academy_course = grouped.count()
display(df_academy_course)

df_academy_rate = df.loc[:,['학원명','취업률']]
grouped = df_academy_rate.groupby(['학원명'])
df_academy_rate = grouped.mean()
display(df_academy_rate)

final = pd.merge(df_academy_course, df_academy_rate, on='학원명')
display(final)

fig = plt.figure(figsize=(20, 10))   
ax1 = fig.add_subplot(1, 1, 1)
sns.boxplot(x='과정', y='취업률', data=final, ax=ax1)
plt.show()
plt.close()
```

## 3. 분야별 취업률이 높은 학원

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("hrdcrolling/preprocessing/final.csv", index_col = 0)

df_academy_rank = df.loc[:,['NCS','학원명','취업률','지역1']]
df_academy_rank = df_academy_rank.groupby(['NCS','학원명','지역1']).mean()
display(df_academy_rank)

df_academy_03 = df_academy_rank.loc['24_농림어업'].sort_values(by=['취업률'], ascending=False)
display(df_academy_03)
df_academy_03.reset_index(level=['지역1'], inplace = True)
display(df_academy_03)

fig = plt.figure(figsize=(20, 10))   
ax1 = fig.add_subplot(1, 1, 1)
sns.barplot(x=df_academy_03.index, y='취업률', hue='지역1', data=df_academy_03, ax=ax1)
plt.show()
plt.close()
```
