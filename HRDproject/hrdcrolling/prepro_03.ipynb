```python
import pandas as pd
from konlpy.tag import Okt
from konlpy.utils import pprint
import nltk
from konlpy.tag import Hannanum 
import re
```

```python
df = pd.read_csv("preprocessing/final.csv")
df.dropna(subset=['취업률'], inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)


df_c = df['NCS'].copy()
df_a = df_c.tolist()
df_a = '\n'.join(df_a)

hannanum = Hannanum()

t = Okt()
tokens_ko = hannanum.nouns(df_a)


ko_ = nltk.Text(tokens_ko)  
ko_.vocab()
freq_dic = dict(ko_.vocab())
sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)
sortedDict = sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)

sortlist = []
for i in range(len(sortedDict)) :
    sortlist.append(sortedDict[i][0])
print(sortlist)
```

```python
comparelist = ['자산관리사', '보험사무원', '보험총무사무원']
nonelist = ['자산관리사', '보험사무원', '보험사무원']
courselist = []
for i in range(len(df)) :
    for n in range(len(comparelist)) :
        if comparelist[n] in df_c[i] :
            course = nonelist[n]
            break;
        else :
            course = None
            
    courselist.append(course)

df['과정'] = courselist
display(df)

print(df.isnull().sum())
df_p = df['과정'].tolist()
for i in range(len(df_p)) :
    if not(df_p[i]) :
        print(i, df.iloc[i]['과정명'])
        
df.to_csv("preprocessing/03.csv")
```
