```python
import pandas as pd
from konlpy.tag import Okt
from konlpy.utils import pprint
import nltk
from konlpy.tag import Hannanum 
import re
```

```python
df = pd.read_csv("10_영업판매.csv")
df.dropna(subset=['취업률'], inplace=True)
df.reset_index(drop=True, inplace=True)
df.drop('Unnamed: 0', axis=1, inplace=True)
df['NCS분야'] = "10_영업판매"

df_c = df['과정명'].copy()
df_a = df_c.tolist()
df_a = '\n'.join(df_a)

hannanum = Hannanum()

t = Okt()
tokens_ko = hannanum.nouns(df_a)


ko_ = nltk.Text(tokens_ko)  
ko_.vocab()
freq_dic = dict(ko_.vocab())
sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)
sortedDict = sorted(freq_dic.items(), key=lambda x: x[1], reverse=True)

sortlist = []
for i in range(len(sortedDict)) :
    sortlist.append(sortedDict[i][0])
print(sortlist)
```

```python
comparelist = ['제작', '구축', '운영', '판매', '셀러', '관리', '상세페이지', '웹디자인']
fixlist = ['쇼핑몰제작', '쇼핑몰제작', '쇼핑몰운영', '쇼핑몰운영', '쇼핑몰운영', '쇼핑몰관리', '쇼핑몰관리', '쇼핑몰관리']
courselist = []
for i in range(len(df)) :
    for n in range(len(comparelist)) :
        if comparelist[n] in df_c[i] :
            course = fixlist[n]
            break;
        else :
            course = None
            
    courselist.append(course)

df['과정'] = courselist
# df['과정'].replace('쇼핑몰', 'e-비지니스', inplace=True)
# df['과정'].replace('오픈마켓', 'e-비지니스', inplace=True)
# df['과정'].replace('스마트스토어', 'e-비지니스', inplace=True)
# df['과정'].replace('글로벌셀러', 'e-비지니스', inplace=True)
display(df)

print(df.isnull().sum())
df_p = df['과정'].tolist()
for i in range(len(df_p)) :
    if not(df_p[i]) :
        print(i, df.iloc[i]['과정명'])
        
df.to_csv("preprocessing/10.csv")
```
