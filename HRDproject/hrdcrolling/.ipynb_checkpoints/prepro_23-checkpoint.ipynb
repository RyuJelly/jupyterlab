```python
import pandas as pd
import numpy as np
from konlpy.tag import Okt
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Hannanum

df = pd.read_csv("23_환경에너지안전.csv", index_col = 0)
# 취업률 없는행 삭제
df = df.dropna(subset=['취업률'], how='any', axis=0)
df.reset_index(drop=True, inplace=True)
display(df)
```

```python
comparelist = ['경영','평가','산업안전', '비파괴', '폐기물', '태양', '신재생', '광물','보건','관리', '컨설팅', '수질측정', '처리']
fixlist = ['환경경영','환경평가','산업안전관리', '비파괴검사', '폐기물관리', '태양광발전설비',
           '신재생에너지발전설비', '광물개발','환경보건관리','환경보건관리','환경보건관리', '수질분석가', '폐수처리기술자']

courselist = []
for i in range(len(df)) :
    for n in range(len(comparelist)) :
        if comparelist[n] in df_c[i] :
            course = fixlist[n]
            break;
        else :
            course = None
    courselist.append(course)

df['NCS'] = "23_환경에너지안전"
df['과정'] = courselist
display(df)


df.to_csv("preprocessing/23.csv")
```
